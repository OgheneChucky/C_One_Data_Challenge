```python
#Remove duplicate Data
tickets = tickets.drop_duplicates()
```

```python
tickets.shape
```

```python

#Remove unnecessary columns

#Remove columns in tickets data
tickets = tickets.drop(['ITIN_ID', 'YEAR', 'QUARTER', 'ORIGIN_COUNTRY', 'ORIGIN_STATE_NM'], axis = 1)

#Remove columns in flights data
flights = flights.drop(['FL_DATE', 'TAIL_NUM', 'OP_CARRIER_FL_NUM', 'ORIGIN_AIRPORT_ID', 'ORIGIN_CITY_NAME','DEST_AIRPORT_ID','DEST_CITY_NAME', 'DEST_AIRPORT_ID',
    'DEST_CITY_NAME', 'DEP_DELAY', 'ARR_DELAY','AIR_TIME', 'DISTANCE', 'OCCUPANCY_RATE'], axis = 1)
```

```python
flights.head()
```

```python
tickets.head()
```

```python
#Remove cancelled flights
flights = flights[flights['CANCELLED'] == 0 ]

#Remove one way flights
tickets = tickets[tickets['ROUNDTRIP'] == 1]
```

```python
#Create new origin to Destination column tickets data
tickets['ORG_DEST'] = tickets['ORIGIN'] + '_' + tickets['DESTINATION']
tickets.head()
```

```python
#Create new origin to Destination column flights data
flights['ORG_DEST'] = flights['ORIGIN'] + '_' + flights['DESTINATION']
flights.head()
```

```python

tickets['REPORTING_CARRIER'].unique()
```

```python
flights['OP_CARRIER'].unique()
```

```python
flights['ORIGIN'].unique()
```

```python
tickets['ORIGIN'].unique()
```

```python
#merge the ticket and flights data
merged_flights_tickets = flights.merge(tickets, left_on=['ORG_DEST','OP_CARRIER'], right_on=['ORG_DEST','REPORTING_CARRIER'], how='inner')
merged_flights_tickets.head()
```

```python
# Remove Duplicates from Merged Flights & Tickets Data
merged_flights_tickets_unique = merged_flights_tickets.drop_duplicates()

# Save the unique data to a CSV file
merged_flights_tickets_unique.to_csv('merged_flights_tickets.csv', index=False)
```

```python
# Show  the busiest roundtrips 
roundtrip_counts = merged_flights_tickets_unique.groupby('ORG_DEST')['ROUNDTRIP'].count().reset_index(name = 'ROUNDTRIP_COUNT')
roundtrip_counts_sorted = roundtrip_counts.sort_values('ROUNDTRIP_COUNT', ascending=False)
print(roundtrip_counts_sorted.head(10))
```
